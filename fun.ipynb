```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt

import torch
from torchvision import transforms
```

```python
def preprocess_image(img):
    transform = transforms.Compose([
        transforms.ToTensor(),
        #transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
        transforms.Normalize((0, 0, 0), (1.0, 1.0, 1.0))
    ])
    img = transform(img)
    return img
```
